# Анализ результатов

```python
import os

import numpy as np
import matplotlib.pyplot as plt

from matplotlib.image import imread
from IPython.display import Image, display
```

```python
labels = ['LogReg','RFC','CatBoost']
score =  [0.7329,0.701,0.761]

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots()
ax2 = ax.twinx()
rects1 = ax.bar(x - width/2, score, width, label='score', color='lightgreen')

ax.set_title('Рейтинг моделей по метрикe Roc-auc')
ax.set_xticks(x - width/2.5)
ax.set_xticklabels(labels)
ax.set_yticks(np.arange(0.4, 0.8, 0.01))
ax.set_ylim([0.5, 0.8])


for rect in rects1:
    height = rect.get_height()
    ax.annotate('{}'.format(f'{height:0.3f}'), 
                xy=(rect.get_x() + rect.get_width() / 2, height), 
                xytext=(0, 2), textcoords='offset points', 
                ha='center', va='bottom')
    
fig.tight_layout()
plt.show();

plt.savefig('res_model_stats.png');
```

### Заключение
По результатам эксперимента построена модель (CatboostClassifier) с метрикой  Roc-auc = 0.761 > 0.75.
